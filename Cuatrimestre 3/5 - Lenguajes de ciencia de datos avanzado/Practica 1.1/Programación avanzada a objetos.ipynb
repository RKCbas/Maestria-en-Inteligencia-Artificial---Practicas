### Ejercicio 1: Sistema de Vehículos con Clases Abstractas y Polimorfismo

Se diseña un sistema de vehículos utilizando el módulo `abc` de Python:

- `Vehiculo`: clase abstracta base con atributos comunes (`marca`, `modelo`, `anio`, `kilometraje`) y los métodos abstractos `calcular_costo_operacion()` y `describir()`.
- `Auto`, `Motocicleta`, `Camion`: subclases concretas que implementan los métodos abstractos con lógicas de costo diferenciadas.
- El **polimorfismo** se demuestra llamando al mismo método sobre una lista de objetos de distintos tipos, obteniendo resultados específicos a cada clase.

```python
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    """Clase abstracta base que define la interfaz común para todos los vehículos."""

    def __init__(self, marca: str, modelo: str, anio: int):
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self._kilometraje = 0.0

    @property
    def kilometraje(self) -> float:
        return self._kilometraje

    def conducir(self, km: float) -> None:
        """Acumula kilómetros recorridos. Valida que el valor sea positivo."""
        if km <= 0:
            raise ValueError("Los kilómetros deben ser un valor positivo.")
        self._kilometraje += km

    @abstractmethod
    def calcular_costo_operacion(self) -> float:
        """Calcula el costo total de operación según tipo de vehículo."""
        pass

    @abstractmethod
    def describir(self) -> str:
        """Retorna una descripción detallada del vehículo."""
        pass

    def __str__(self) -> str:
        return f"{self.anio} {self.marca} {self.modelo}"


class Auto(Vehiculo):
    """Automóvil de pasajeros — costo de $0.15 por km."""

    def __init__(self, marca: str, modelo: str, anio: int, num_puertas: int = 4):
        super().__init__(marca, modelo, anio)
        self.num_puertas = num_puertas

    def calcular_costo_operacion(self) -> float:
        return self._kilometraje * 0.15

    def describir(self) -> str:
        return (f"Auto:  {self} | Puertas: {self.num_puertas} | "
                f"Km: {self._kilometraje:>8.1f} | Costo: ${self.calcular_costo_operacion():.2f}")


class Motocicleta(Vehiculo):
    """Motocicleta — costo de $0.08 por km."""

    def __init__(self, marca: str, modelo: str, anio: int, cilindrada: int):
        super().__init__(marca, modelo, anio)
        self.cilindrada = cilindrada

    def calcular_costo_operacion(self) -> float:
        return self._kilometraje * 0.08

    def describir(self) -> str:
        return (f"Moto:  {self} | Cilindrada: {self.cilindrada}cc | "
                f"Km: {self._kilometraje:>8.1f} | Costo: ${self.calcular_costo_operacion():.2f}")


class Camion(Vehiculo):
    """Camión de carga — costo variable según tonelaje."""

    def __init__(self, marca: str, modelo: str, anio: int, capacidad_ton: float):
        super().__init__(marca, modelo, anio)
        self.capacidad_ton = capacidad_ton

    def calcular_costo_operacion(self) -> float:
        # Factor adicional por tonelaje de carga
        return self._kilometraje * 0.35 * (1 + self.capacidad_ton / 20)

    def describir(self) -> str:
        return (f"Camión:{self} | Capacidad: {self.capacidad_ton}t | "
                f"Km: {self._kilometraje:>8.1f} | Costo: ${self.calcular_costo_operacion():.2f}")


# --- Pruebas ---
print("=== Ejercicio 1: Sistema de Vehículos ===\n")

flota: list[Vehiculo] = [
    Auto("Toyota", "Corolla", 2022, 4),
    Motocicleta("Honda", "CBR600", 2021, 600),
    Camion("Volvo", "FH16", 2020, 18.0),
]

for vehiculo, km in zip(flota, [150, 200, 100]):
    vehiculo.conducir(km)
    print(vehiculo.describir())

# Polimorfismo: mismo método, comportamientos distintos según el tipo real
print("\n--- Demostración de polimorfismo ---")
for v in flota:
    print(f"  {type(v).__name__:12s} → calcular_costo_operacion() = ${v.calcular_costo_operacion():.2f}")
```

### Ejercicio 2: Patrón de Diseño Observer

El **Patrón Observer** (GoF) permite que múltiples objetos "observadores" sean notificados automáticamente cuando un "sujeto" cambia de estado, sin crear dependencias rígidas entre ellos.

- `ObservadorBase`: clase abstracta con el método `actualizar()` que todos los observadores deben implementar.
- `Sujeto`: mixin reutilizable que gestiona la lista de observadores (`suscribir`, `desuscribir`, `notificar`).
- `SensorTemperatura`: sujeto concreto que usa un setter de propiedad para disparar notificaciones automáticamente.
- `AlertaConsola` y `AlertaRegistro`: observadores concretos con comportamientos independientes.

```python
from abc import ABC, abstractmethod
from datetime import datetime
from typing import List

class ObservadorBase(ABC):
    """Interfaz abstracta para todos los observadores del Patrón Observer."""

    @abstractmethod
    def actualizar(self, temperatura: float, fuente: str) -> None:
        """Recibe la notificación de un cambio de estado."""
        pass


class Sujeto:
    """Mixin reutilizable que gestiona suscriptores y emite notificaciones."""

    def __init__(self):
        self._observadores: List[ObservadorBase] = []

    def suscribir(self, observador: ObservadorBase) -> None:
        if observador not in self._observadores:
            self._observadores.append(observador)

    def desuscribir(self, observador: ObservadorBase) -> None:
        self._observadores.remove(observador)

    def notificar(self, temperatura: float, fuente: str) -> None:
        for obs in self._observadores:
            obs.actualizar(temperatura, fuente)


class SensorTemperatura(Sujeto):
    """Sujeto concreto: emite notificaciones cada vez que cambia la temperatura."""

    def __init__(self, nombre: str):
        super().__init__()
        self.nombre = nombre
        self._temperatura = 0.0

    @property
    def temperatura(self):
        return self._temperatura

    @temperatura.setter
    def temperatura(self, valor: float):
        self._temperatura = valor
        self.notificar(valor, self.nombre)  # Notifica automáticamente al cambiar


class AlertaConsola(ObservadorBase):
    """Observador concreto: muestra alertas en consola según umbral configurado."""

    def __init__(self, umbral: float = 75.0):
        self.umbral = umbral

    def actualizar(self, temperatura: float, fuente: str) -> None:
        estado = "¡CRITICA!" if temperatura >= self.umbral else "normal"
        print(f"  [Consola] {fuente}: {temperatura:.1f}°C — Estado: {estado}")


class AlertaRegistro(ObservadorBase):
    """Observador concreto: acumula un historial de lecturas para análisis."""

    def __init__(self):
        self.historial: list = []

    def actualizar(self, temperatura: float, fuente: str) -> None:
        self.historial.append({
            "timestamp": datetime.now().strftime("%H:%M:%S"),
            "fuente": fuente,
            "temperatura": temperatura,
        })

    def mostrar_historial(self):
        print("\n  [Registro] Historial completo:")
        for e in self.historial:
            print(f"    {e['timestamp']} | {e['fuente']}: {e['temperatura']}°C")


# --- Pruebas ---
print("=== Patrón Observer: Monitoreo de Temperatura ===\n")

sensor_cpu = SensorTemperatura("CPU")
sensor_gpu = SensorTemperatura("GPU")

alerta = AlertaConsola(umbral=75.0)
registro = AlertaRegistro()

sensor_cpu.suscribir(alerta)
sensor_cpu.suscribir(registro)
sensor_gpu.suscribir(alerta)
sensor_gpu.suscribir(registro)

print("Lecturas CPU:")
for temp in [45.0, 60.0, 78.0, 91.0, 55.0]:
    sensor_cpu.temperatura = temp

print("\nLecturas GPU:")
for temp in [50.0, 72.0, 88.0]:
    sensor_gpu.temperatura = temp

registro.mostrar_historial()
```

### Ejercicio 3: Patrón de Diseño Strategy

El **Patrón Strategy** define una familia de algoritmos intercambiables encapsulados en clases separadas. Aquí se aplica al cálculo de descuentos en un carrito de compras:

- `EstrategiaDescuento`: clase abstracta que define el contrato (`aplicar`, `descripcion`).
- `DescuentoPorcentaje`, `DescuentoFijo`, `DescuentoLeal`: estrategias concretas con distintas reglas de negocio.
- `Carrito`: contexto que delega el cálculo del descuento a la estrategia activa, pudiéndola cambiar en tiempo de ejecución sin alterar su propio código.

```python
from abc import ABC, abstractmethod
from typing import List, Tuple

class EstrategiaDescuento(ABC):
    """Interfaz abstracta para estrategias de descuento (Patrón Strategy)."""

    @abstractmethod
    def aplicar(self, total: float) -> float:
        """Retorna el monto de descuento a aplicar sobre el total."""
        pass

    @abstractmethod
    def descripcion(self) -> str:
        pass


class DescuentoPorcentaje(EstrategiaDescuento):
    """Descuento basado en un porcentaje del total."""

    def __init__(self, porcentaje: float):
        if not (0 < porcentaje <= 100):
            raise ValueError("El porcentaje debe estar entre 0 y 100.")
        self.porcentaje = porcentaje

    def aplicar(self, total: float) -> float:
        return total * (self.porcentaje / 100)

    def descripcion(self) -> str:
        return f"Descuento {self.porcentaje}%"


class DescuentoFijo(EstrategiaDescuento):
    """Descuento de monto fijo (sin superar el total)."""

    def __init__(self, monto: float):
        self.monto = monto

    def aplicar(self, total: float) -> float:
        return min(self.monto, total)

    def descripcion(self) -> str:
        return f"Descuento fijo ${self.monto:.2f}"


class DescuentoLeal(EstrategiaDescuento):
    """Descuento escalonado para clientes frecuentes según el monto total."""

    def aplicar(self, total: float) -> float:
        if total >= 500:
            return total * 0.20
        elif total >= 200:
            return total * 0.10
        elif total >= 100:
            return total * 0.05
        return 0.0

    def descripcion(self) -> str:
        return "Descuento cliente leal (escalonado: 5%/10%/20%)"


class Carrito:
    """Contexto del Patrón Strategy: aplica la estrategia de descuento activa."""

    def __init__(self, estrategia: EstrategiaDescuento = None):
        self._items: List[Tuple[str, float, int]] = []
        self._estrategia = estrategia

    def cambiar_estrategia(self, estrategia: EstrategiaDescuento):
        self._estrategia = estrategia

    def agregar(self, nombre: str, precio: float, cantidad: int = 1):
        self._items.append((nombre, precio, cantidad))

    def subtotal(self) -> float:
        return sum(precio * cant for _, precio, cant in self._items)

    def calcular_total(self) -> float:
        sub = self.subtotal()
        descuento = self._estrategia.aplicar(sub) if self._estrategia else 0.0
        return sub - descuento

    def resumen(self):
        estrategia_nombre = self._estrategia.descripcion() if self._estrategia else "Ninguna"
        print(f"  Estrategia: {estrategia_nombre}")
        for nombre, precio, cant in self._items:
            print(f"    {nombre} x{cant}: ${precio * cant:.2f}")
        sub = self.subtotal()
        desc = self._estrategia.aplicar(sub) if self._estrategia else 0.0
        print(f"  Subtotal: ${sub:.2f} | Descuento: -${desc:.2f} | Total: ${sub - desc:.2f}\n")


# --- Pruebas ---
print("=== Patrón Strategy: Carrito de Compras ===\n")

carrito = Carrito()
carrito.agregar("Laptop", 899.99)
carrito.agregar("Mouse", 29.99, 2)
carrito.agregar("Teclado", 79.99)

for estrategia in [DescuentoPorcentaje(10), DescuentoFijo(50), DescuentoLeal()]:
    carrito.cambiar_estrategia(estrategia)
    carrito.resumen()
```

### Ejercicio 4: Decorador de Tiempo de Ejecución

El decorador `@medir_tiempo` envuelve cualquier función y mide el tiempo que tarda en ejecutarse usando `time.perf_counter()` (alta resolución). Imprime la duración en milisegundos sin modificar el valor de retorno de la función original. El uso de `functools.wraps` asegura que la función decorada conserve su nombre (`__name__`) y documentación (`__doc__`), lo que es esencial para mantener la transparencia del decorador.

```python
import time
import functools

def medir_tiempo(func):
    """
    Decorador que mide y muestra el tiempo de ejecución de una función.
    Usa functools.wraps para preservar el nombre y docstring original.
    """
    @functools.wraps(func)
    def envoltura(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = func(*args, **kwargs)
        fin = time.perf_counter()
        duracion_ms = (fin - inicio) * 1000
        print(f"  [{func.__name__}] Tiempo de ejecución: {duracion_ms:.3f} ms")
        return resultado
    return envoltura


@medir_tiempo
def suma_cuadrados(n: int) -> int:
    """Suma los cuadrados de 1 a n."""
    return sum(i ** 2 for i in range(1, n + 1))


@medir_tiempo
def busqueda_lineal(lista: list, objetivo) -> int:
    """Retorna el índice del objetivo en la lista, o -1 si no existe."""
    for i, val in enumerate(lista):
        if val == objetivo:
            return i
    return -1


@medir_tiempo
def ordenar_burbuja(lista: list) -> list:
    """Ordenamiento burbuja O(n²) — útil para demostrar diferencias de rendimiento."""
    arr = lista[:]
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# --- Pruebas ---
print("=== Decorador: Tiempo de Ejecución ===\n")

resultado = suma_cuadrados(100_000)
print(f"  Resultado: {resultado}\n")

datos = list(range(10_000, 0, -1))
idx = busqueda_lineal(datos, 5_000)
print(f"  Índice encontrado: {idx}\n")

ordenado = ordenar_burbuja(list(range(500, 0, -1)))
print(f"  Primeros 5 elementos: {ordenado[:5]}")
```

### Ejercicio 5: Decorador de Reintentos Automáticos

El decorador `@reintentar` es un **decorador de fábrica**: recibe parámetros de configuración y devuelve el decorador real. Intenta ejecutar la función hasta `max_intentos` veces cuando ocurre una excepción del tipo especificado en `excepciones`, esperando `espera` segundos entre cada intento. Si se agotan los intentos, relanza la última excepción. Es ideal para operaciones que pueden fallar transitoriamente, como llamadas a servicios externos o lectura de archivos en red.

```python
import time
import functools

def reintentar(max_intentos: int = 3, espera: float = 0.5, excepciones: tuple = (Exception,)):
    """
    Decorador de fábrica que reintenta una función ante excepciones específicas.

    Args:
        max_intentos: número máximo de ejecuciones.
        espera: segundos de pausa entre intentos.
        excepciones: tipos de excepción que activan el reintento.
    """
    def decorador(func):
        @functools.wraps(func)
        def envoltura(*args, **kwargs):
            for intento in range(1, max_intentos + 1):
                try:
                    resultado = func(*args, **kwargs)
                    if intento > 1:
                        print(f"  [{func.__name__}] Éxito en el intento {intento}.")
                    return resultado
                except excepciones as e:
                    print(f"  [{func.__name__}] Intento {intento}/{max_intentos} fallido: {e}")
                    if intento < max_intentos:
                        time.sleep(espera)
                    else:
                        print(f"  [{func.__name__}] Se agotaron los intentos.")
                        raise
        return envoltura
    return decorador


# Contador de llamadas encapsulado para evitar estado global
_estado = {"llamadas": 0}

@reintentar(max_intentos=4, espera=0.05, excepciones=(ConnectionError, TimeoutError))
def consultar_api_externa() -> str:
    """Simula una llamada a API que falla las primeras 3 veces."""
    _estado["llamadas"] += 1
    if _estado["llamadas"] < 4:
        raise ConnectionError(f"Sin respuesta (intento {_estado['llamadas']})")
    return "{'status': 'ok', 'data': [1, 2, 3]}"


@reintentar(max_intentos=2, espera=0.0)
def division_segura(a: float, b: float) -> float:
    """División que falla cuando b == 0 (demuestra agotamiento de intentos)."""
    if b == 0:
        raise ValueError("División por cero.")
    return a / b


# --- Pruebas ---
print("=== Decorador: Reintentos Automáticos ===\n")

print("Caso 1: API que eventualmente responde:")
respuesta = consultar_api_externa()
print(f"  Respuesta final: {respuesta}\n")

print("Caso 2: Función que siempre falla (agota intentos):")
try:
    division_segura(10, 0)
except ValueError as e:
    print(f"  Error final capturado: {e}")
```

### Ejercicio 6: Decorador de Validación de Tipos

El decorador `@validar_tipos` usa las anotaciones de tipo (`type hints`) de la función decorada para verificar en tiempo de ejecución que los argumentos recibidos sean del tipo correcto. Emplea `inspect.signature` para asociar argumentos posicionales con sus nombres, y `typing.get_type_hints` para obtener los tipos esperados. Si un argumento no cumple el tipo esperado, lanza un `TypeError` con un mensaje descriptivo que indica qué se recibió y qué se esperaba.

```python
import functools
import inspect
from typing import get_type_hints

def validar_tipos(func):
    """
    Decorador que valida los tipos de los argumentos usando las anotaciones de tipo.
    Utiliza inspect y get_type_hints para introspección en tiempo de ejecución.
    """
    hints = get_type_hints(func)
    sig = inspect.signature(func)

    @functools.wraps(func)
    def envoltura(*args, **kwargs):
        argumentos = sig.bind(*args, **kwargs)
        argumentos.apply_defaults()

        for nombre, valor in argumentos.arguments.items():
            if nombre in hints:
                tipo_esperado = hints[nombre]
                # Saltar tipos genéricos complejos (List[int], etc.)
                if hasattr(tipo_esperado, '__origin__'):
                    continue
                if not isinstance(valor, tipo_esperado):
                    raise TypeError(
                        f"Argumento '{nombre}' debe ser {tipo_esperado.__name__}, "
                        f"se recibió {type(valor).__name__} (valor: {valor!r})."
                    )
        return func(*args, **kwargs)
    return envoltura


@validar_tipos
def calcular_imc(peso: float, altura: float, nombre: str) -> float:
    """Calcula el Índice de Masa Corporal."""
    imc = peso / (altura ** 2)
    print(f"  IMC de {nombre}: {imc:.2f}")
    return imc


@validar_tipos
def crear_perfil(nombre: str, edad: int, activo: bool = True) -> dict:
    """Crea un diccionario de perfil de usuario."""
    return {"nombre": nombre, "edad": edad, "activo": activo}


# --- Pruebas ---
print("=== Decorador: Validación de Tipos ===\n")

print("Caso 1: Tipos correctos:")
calcular_imc(70.5, 1.75, "Ana")
perfil = crear_perfil("Carlos", 28)
print(f"  Perfil: {perfil}\n")

print("Caso 2: Tipo incorrecto en 'peso' (str en lugar de float):")
try:
    calcular_imc("setenta", 1.75, "Luis")
except TypeError as e:
    print(f"  Error: {e}\n")

print("Caso 3: Tipo incorrecto en 'edad' (float en lugar de int):")
try:
    crear_perfil("María", 25.5)
except TypeError as e:
    print(f"  Error: {e}")
```

### Descripción — Funciones y Decoradores Personalizados

Los tres decoradores implementados ilustran cómo Python permite extender el comportamiento de funciones de forma declarativa y reutilizable. En el **Ejercicio 4**, `@medir_tiempo` utiliza `time.perf_counter()` para cronometrar la ejecución con precisión de microsegundos. El uso de `functools.wraps` es fundamental: sin él, la función resultante perdería su `__name__`, `__doc__` y `__module__`, rompiendo herramientas de introspección como `help()`.

En el **Ejercicio 5**, `@reintentar` es un **decorador de fábrica** (decorador que recibe parámetros), lo que introduce una capa extra de anidamiento: `decorador(func)` es el verdadero decorador y `envoltura(*args, **kwargs)` es la función resultante. Los parámetros `max_intentos`, `espera` y `excepciones` quedan capturados en el closure, permitiendo reutilizar el decorador con distintas configuraciones de forma concisa, por ejemplo `@reintentar(max_intentos=5, espera=1.0, excepciones=(IOError,))`.

En el **Ejercicio 6**, `@validar_tipos` emplea `inspect.signature` y `typing.get_type_hints` para introspección en tiempo de ejecución, demostrando el uso avanzado de `*args` y `**kwargs` combinado con `sig.bind()` para asociar argumentos posicionales con sus nombres. Este decorador aporta una capa de seguridad de tipos en tiempo de ejecución similar a lo que ofrece mypy en tiempo de análisis estático, útil en APIs internas donde los errores de tipo deben detectarse con mensajes claros.
