La columna **key** coincide con la columna **key** en `reviews.csv`.

# Actividades

Con la información anterior, realice lo siguiente:

### CONFIGURACIÓN INICIAL

```r
# Cargar librerías necesarias
library(stringr)
library(dplyr)
library(tidyr)

# Cargar los datos
reviews <- read.csv("./datos/reviews.csv", stringsAsFactors = FALSE)
products <- read.csv("./datos/products.csv", stringsAsFactors = FALSE)

# Ver estructura de los datos
str(reviews)
head(reviews)
```

### 1.- Identifique las **10 observaciones con el nombre de usuario más largo**.

```r
# Calcular la longitud de cada nombre de usuario
reviews <- reviews %>%
  mutate(author_length = str_length(author))

# Identificar las 10 observaciones con nombres más largos
top_10_longest <- reviews %>%
  arrange(desc(author_length)) %>%
  select(author, author_length, date, stars) %>%
  head(10)

print(top_10_longest)
```

### 2.- Transforme los nombres de usuarios **author** y las reseñas **text** a **minúsculas**.

```r
# Transformar author y text a minúsculas
reviews <- reviews %>%
  mutate(
    author_lower = str_to_lower(author),
    text_lower = str_to_lower(text)
  )

# Verificar algunos ejemplos
reviews %>%
  select(author, author_lower, text, text_lower) %>%
  head(3)
```

### 3.- Transforme la variable **title** de tal forma que **cada una de las palabras inicie con mayúscula**.

```r
# Transformar title a formato título (cada palabra con mayúscula inicial)
reviews <- reviews %>%
  mutate(title_titlecase = str_to_title(title))

# Verificar algunos ejemplos
reviews %>%
  select(title, title_titlecase) %>%
  head(10)
```

### 4.- Transforme el contenido de las reseñas de tal forma que **solo la primera palabra sea mayúscula**.

```r
# Transformar text para que solo la primera palabra sea mayúscula
reviews <- reviews %>%
  mutate(text_sentence = str_to_sentence(text))

# Verificar algunos ejemplos
reviews %>%
  select(text, text_sentence) %>%
  head(3)
```

### 5.- Explique **cómo puede afectar la normalización de texto al análisis de datos de texto**.

1. CONSISTENCIA: Evita contar 'Delicious', 'delicious' y 'DELICIOUS' como palabras diferentes.

2. REDUCCIÓN DE DIMENSIONALIDAD: Disminuye el número de términos únicos en el diccionario.

3. MEJORA EN BÚSQUEDAS: Facilita encontrar patrones sin importar el formato original.

4. ANÁLISIS DE SENTIMIENTOS: Permite identificar palabras clave de manera más efectiva.

5. PÉRDIDA DE INFORMACIÓN: Puede perder matices como énfasis (MAYÚSCULAS) o nombres propios.

6. COMPARABILIDAD: Facilita comparaciones entre diferentes fuentes de texto.

### 6.- Identifique las **reseñas positivas** usando como criterio la aparición de la palabra:

- `delicious`

Determine en cuántas reseñas se usó la palabra.

Haga lo mismo para:

- `good`
- `excellent`

¿Se observa algo de interés?

```r
# Usar text_lower para búsquedas (creada en problema 2)
reviews$text_lower <- str_to_lower(reviews$text)

# Palabra: "delicious"
delicious_reviews <- reviews %>%
  filter(str_detect(text_lower, "delicious"))

cat("Reseñas con 'delicious':", nrow(delicious_reviews), "\n")

# Palabra: "good"
good_reviews <- reviews %>%
  filter(str_detect(text_lower, "good"))

cat("Reseñas con 'good':", nrow(good_reviews), "\n")

# Palabra: "excellent"
excellent_reviews <- reviews %>%
  filter(str_detect(text_lower, "excellent"))

cat("Reseñas con 'excellent':", nrow(excellent_reviews), "\n")

# Resumen
positive_summary <- data.frame(
  Palabra = c("delicious", "good", "excellent"),
  Cantidad = c(
               nrow(delicious_reviews),
               nrow(good_reviews),
               nrow(excellent_reviews))
)

print(positive_summary)
```

### 7.- De forma similar al ejercicio anterior, identifique las **reseñas negativas** con el uso de la palabra:

- `bad`

Haga lo mismo para la palabra:

- `worst`

¿Qué otras palabras podría utilizar?

```r
# Palabra: "bad"
bad_reviews <- reviews %>%
  filter(str_detect(text_lower, "bad"))

cat("Reseñas con 'bad':", nrow(bad_reviews), "\n")

# Palabra: "worst"
worst_reviews <- reviews %>%
  filter(str_detect(text_lower, "worst"))

cat("Reseñas con 'worst':", nrow(worst_reviews), "\n")

# Otras palabras negativas
otras_negativas <- c("terrible", "awful", "disgusting",
                     "hate", "disappointing", "disappointed")

resultados_negativos <- data.frame(
  Palabra = c("bad", "worst", otras_negativas),
  Cantidad = sapply(c("bad", "worst", otras_negativas), function(palabra) {
    sum(str_detect(reviews$text_lower, palabra))
  })
)

print(resultados_negativos)
```

### 8.- Cree un **diccionario con todas las palabras de cada reseña** y **contabilícelas**.

Exporte un archivo **.CSV** con este diccionario.

Para el análisis de los textos, considere alguna o algunas otras palabras.

```r
# Extraer todas las palabras de todas las reseñas
todas_palabras <- reviews %>%
  pull(text_lower) %>%
  str_split("\\s+") %>%  # Dividir por espacios
  unlist() %>%
  str_remove_all("[^a-z']") %>%  # Remover puntuación excepto apóstrofes
  str_remove_all("^'|'$")  # Remover apóstrofes al inicio/final

# Filtrar palabras vacías
todas_palabras <- todas_palabras[todas_palabras != ""]

# Crear diccionario con conteo
diccionario <- data.frame(palabra = todas_palabras) %>%
  group_by(palabra) %>%
  summarise(frecuencia = n()) %>%
  arrange(desc(frecuencia))

# Ver las palabras más frecuentes
head(diccionario, 20)

# Exportar a CSV
write.csv(diccionario, "./datos/diccionario_palabras.csv", row.names = FALSE)

cat("Diccionario exportado con", nrow(diccionario), "palabras únicas\n")
```

### 9.- Determine si se observa **algún sentimiento específico de los usuarios con respecto a los sabores de los helados**.

```r
# Análisis de sentimientos básico
palabras_positivas <- c("delicious", "good", "excellent", "amazing", "love",
                        "great", "perfect", "awesome", "best", "favorite")

palabras_negativas <- c("bad", "worst", "terrible", "awful", "hate",
                        "disappointing", "poor", "horrible", "disgusting")

# Contar palabras positivas y negativas por reseña
reviews <- reviews %>%
  mutate(
    positivas_count = sapply(text_lower, function(x) {
      sum(str_count(x, palabras_positivas))
    }),
    negativas_count = sapply(text_lower, function(x) {
      sum(str_count(x, palabras_negativas))
    }),
    sentimiento = case_when(
      positivas_count > negativas_count ~ "Positivo",
      negativas_count > positivas_count ~ "Negativo",
      TRUE ~ "Neutral"
    )
  )

# Resumen de sentimientos
tabla_sentimientos <- reviews %>%
  group_by(sentimiento) %>%
  summarise(
    cantidad = n(),
    promedio_estrellas = mean(stars, na.rm = TRUE)
  )

print(tabla_sentimientos)

# Análisis por estrellas
sentimiento_por_estrellas <- reviews %>%
  group_by(stars, sentimiento) %>%
  summarise(cantidad = n(), .groups = "drop")

print(sentimiento_por_estrellas)
```

### 10.- Determine **cuáles son los sabores más populares**.

¿Será algún ingrediente particular?

```r
# Contar reseñas por sabor (key)
sabores_populares <- reviews %>%
  group_by(key) %>%
  summarise(
    num_reviews = n(),
    promedio_estrellas = mean(stars, na.rm = TRUE)
  ) %>%
  arrange(desc(num_reviews))

# Unir con información de productos
if (exists("products")) {
  sabores_populares <- sabores_populares %>%
    left_join(products %>% select(key, name, rating), by = "key")
}

head(sabores_populares, 10)

# Analizar ingredientes más mencionados
if (exists("products")) {
  # Extraer todos los ingredientes
  ingredientes_texto <- products %>%
    pull(ingredients) %>%
    str_to_lower() %>%
    paste(collapse = ", ")

  # Ingredientes comunes a buscar
  ingredientes_buscar <- c("chocolate", "vanilla", "caramel",
                           "peanut", "cookie", "brownie", "fudge",
                           "cherry", "strawberry", "mint")

  ingredientes_count <- data.frame(
    ingrediente = ingredientes_buscar,
    frecuencia = sapply(ingredientes_buscar, function(ing) {
      sum(str_detect(products$ingredients, regex(ing, ignore_case = TRUE)))
    })
  ) %>%
    arrange(desc(frecuencia))

  print(ingredientes_count)
}
```

### 11.- Identifique **en cuántas reseñas se mencionó el chocolate más de tres veces**.

```r
# Contar menciones de "chocolate" en cada reseña
reviews <- reviews %>%
  mutate(chocolate_count = str_count(text_lower, "chocolate"))

# Filtrar reseñas con más de 3 menciones
chocolate_mas_3 <- reviews %>%
  filter(chocolate_count > 3)

cat("Número de reseñas que mencionan 'chocolate' más de 3 veces:",
    nrow(chocolate_mas_3), "\n")

# Ver distribución
table(reviews$chocolate_count)

# Ejemplos de reseñas con muchas menciones
reviews %>%
  filter(chocolate_count > 3) %>%
  select(author, chocolate_count, stars, text) %>%
  head(3)
```

### 12.- Separe la variable **date** en tres columnas:

- año
- mes
- día

Responda lo siguiente:

- ¿En qué **año** se escribieron más reseñas?
- ¿En qué **mes** se escribieron más reseñas?

```r
# Separar la fecha en columnas
reviews <- reviews %>%
  separate(date, into = c("year", "month", "day"),
           sep = "-", remove = FALSE, convert = TRUE)

# Ver resultado
reviews %>%
  select(date, year, month, day) %>%
  head(10)

# ¿En qué año se escribieron más reseñas?
reviews_por_year <- reviews %>%
  group_by(year) %>%
  summarise(cantidad = n()) %>%
  arrange(desc(cantidad))

cat("\nReseñas por año:\n")
print(reviews_por_year)

cat("\nAño con más reseñas:", reviews_por_year$year[1],
    "con", reviews_por_year$cantidad[1], "reseñas\n")

# ¿En qué mes se escribieron más reseñas?
reviews_por_mes <- reviews %>%
  group_by(month) %>%
  summarise(cantidad = n()) %>%
  arrange(desc(cantidad))

cat("\nReseñas por mes:\n")
print(reviews_por_mes)

cat("\nMes con más reseñas:", reviews_por_mes$month[1],
    "con", reviews_por_mes$cantidad[1], "reseñas\n")
```
