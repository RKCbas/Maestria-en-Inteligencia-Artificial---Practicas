# Graficación con Python

## Objetivo

El objetivo de esta actividad es desarrollar habilidades prácticas en la visualización de datos utilizando matplotlib y seaborn. Los alumnos aprenderán a crear gráficos de líneas, dispersión, barras, mapas de calor y más, mientras manipulan datasets reales como Iris, diabetes, vinos, dígitos, Boston, tips, Titanic y flights. A través de estos ejercicios, los alumnos comprenderán cómo cargar, procesar y visualizar datos de manera efectiva, mejorando su capacidad para explorar y comunicar hallazgos de datos.
