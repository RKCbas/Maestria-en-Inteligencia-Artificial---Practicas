### Ejercicio 1

 - Cree un gráfico de línea para sin(x) en el rango de 0 a 10.
 - Agregue etiquetas a los ejes x e y, y añada un título al gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2, color='blue')

# Etiquetas y título
plt.xlabel('x', fontsize=12)
plt.ylabel('sin(x)', fontsize=12)
plt.title('Gráfico de sin(x) en el rango [0, 10]', fontsize=14, fontweight='bold')

# Mejorar visualización
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)

plt.show()
```

### Ejercicio 2

 - Cree una figura con cuatro subplots (2 × 2). En cada subplot, grafique sin(x), cos(x), tan(x) y exp(x) respectivamente en el rango de 0 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos
x = np.linspace(0, 10, 100)

# Crear figura con 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: sin(x)
axes[0, 0].plot(x, np.sin(x), color='blue', linewidth=2)
axes[0, 0].set_title('sin(x)', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].grid(True, alpha=0.3)

# Subplot 2: cos(x)
axes[0, 1].plot(x, np.cos(x), color='green', linewidth=2)
axes[0, 1].set_title('cos(x)', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].grid(True, alpha=0.3)

# Subplot 3: tan(x)
axes[1, 0].plot(x, np.tan(x), color='red', linewidth=2)
axes[1, 0].set_title('tan(x)', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('tan(x)')
axes[1, 0].set_ylim(-10, 10)  # Limitar eje y para mejor visualización
axes[1, 0].grid(True, alpha=0.3)

# Subplot 4: exp(x)
axes[1, 1].plot(x, np.exp(x), color='purple', linewidth=2)
axes[1, 1].set_title('exp(x)', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('exp(x)')
axes[1, 1].grid(True, alpha=0.3)

# Ajustar espaciado entre subplots
plt.tight_layout()
plt.show() 
```

### Ejercicio 3

 - Genere datos aleatorios y cree un gráfico de dispersión donde el color de los puntos esté determinado por otra variable aleatoria.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear generador de números aleatorios
rng = np.random.default_rng(42)  # Para reproducibilidad
n_puntos = 100

x = rng.random(n_puntos) * 10
y = rng.random(n_puntos) * 10
colores = rng.random(n_puntos)  # Variable que determina el color

# Crear gráfico de dispersión
plt.figure(figsize=(10, 7))
scatter = plt.scatter(x, y, c=colores, cmap='viridis', s=100, alpha=0.7, edgecolors='black')

# Agregar barra de color
plt.colorbar(scatter, label='Valor de color')

# Etiquetas y título
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.title('Gráfico de Dispersión con Colores Aleatorios', fontsize=14, fontweight='bold')

# Mejorar visualización
plt.grid(True, alpha=0.3)

plt.show()
```

### Ejercicio 4

 - Genere datos aleatorios y cree un gráfico de dispersión donde el tamaño de los puntos esté determinado por otra variable aleatoria.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear generador de números aleatorios
rng = np.random.default_rng(seed=42)

# Generar datos aleatorios
n_puntos = 100

x = rng.random(n_puntos) * 10
y = rng.random(n_puntos) * 10
tamaños = rng.random(n_puntos) * 500  # Variable que determina el tamaño

# Crear gráfico de dispersión
plt.figure(figsize=(10, 7))
plt.scatter(x, y, s=tamaños, alpha=0.5, c='steelblue', edgecolors='black', linewidth=0.5)

# Etiquetas y título
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.title('Gráfico de Dispersión con Tamaños Aleatorios', fontsize=14, fontweight='bold')

# Mejorar visualización
plt.grid(True, alpha=0.3)

plt.show()
```

### Ejercicio 5

 - Cargue el dataset Iris utilizando sklearn.datasets. 

 - Cree un gráfico de dispersión mostrando las características sepal length y sepal width. 

 - Use colores diferentes para cada una de las tres especies de flores.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
nombres_especies = iris.target_names

# Extraer características
sepal_length = X[:, 0]
sepal_width = X[:, 1]

# Crear gráfico de dispersión
plt.figure(figsize=(10, 7))

# Graficar cada especie con color diferente
colores = ['red', 'blue', 'green']
for i, especie in enumerate(nombres_especies):
    mask = y == i
    plt.scatter(sepal_length[mask], sepal_width[mask], 
                c=colores[i], label=especie.capitalize(), 
                s=100, alpha=0.6, edgecolors='black', linewidth=0.5)

# Etiquetas y título
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.title('Iris Dataset: Sepal Length vs Sepal Width', fontsize=14, fontweight='bold')

# Leyenda y cuadrícula
plt.legend(title='Especies', fontsize=10)
plt.grid(True, alpha=0.3)

plt.show()
```

### Ejercicio 6

- Cargue el dataset de vinos (load_wine) utilizando sklearn.datasets.

- Genere un histograma de la característica alcohol.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Cargar dataset de vinos
wine = load_wine()
X = wine.data

# Extraer la característica 'alcohol' (primera columna)
alcohol = X[:, 0]

# Crear histograma
plt.figure(figsize=(10, 6))
plt.hist(alcohol, bins=20, color='darkred', alpha=0.7, edgecolor='black')

# Etiquetas y título
plt.xlabel('Alcohol (%)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Distribución del Contenido de Alcohol en Vinos', fontsize=14, fontweight='bold')

# Agregar línea vertical con la media
media = np.mean(alcohol)
plt.axvline(media, color='blue', linestyle='--', linewidth=2, label=f'Media: {media:.2f}%')

# Leyenda y cuadrícula
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, axis='y')

plt.show()
```

### Ejercicio 7

 - Cargue el dataset de vinos.

 - Cree un boxplot de la característica malic acid para cada clase de vino.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Cargar dataset de vinos
wine = load_wine()
X = wine.data
y = wine.target
nombres_clases = wine.target_names

# Extraer la característica 'malic_acid' (segunda columna)
malic_acid = X[:, 1]

# Separar datos por clase
datos_por_clase = [malic_acid[y == i] for i in range(len(nombres_clases))]

# Crear boxplot
plt.figure(figsize=(10, 6))
box = plt.boxplot(datos_por_clase, tick_labels=nombres_clases, patch_artist=True)

# Personalizar colores de las cajas
colores = ['lightcoral', 'lightblue', 'lightgreen']
for patch, color in zip(box['boxes'], colores):
    patch.set_facecolor(color)

# Etiquetas y título
plt.xlabel('Clase de Vino', fontsize=12)
plt.ylabel('Malic Acid', fontsize=12)
plt.title('Distribución de Ácido Málico por Clase de Vino', fontsize=14, fontweight='bold')

# Mejorar visualización
plt.grid(True, alpha=0.3, axis='y')

plt.show()
```

### Ejercicio 8

 - Cargue el dataset de diabetes (load_diabetes) utilizando sklearn.datasets.

 - Cree un gráfico de barras mostrando la media de la característica BMI para cada grupo de

 - progresión de la enfermedad (use la variable objetivo para definir los grupos).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Cargar dataset de diabetes
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target  # Variable objetivo (progresión de la enfermedad)

# Extraer BMI (índice 2)
bmi = X[:, 2]

# Crear grupos basados en cuartiles de la progresión
cuartiles = np.percentile(y, [25, 50, 75])
grupos = np.digitize(y, cuartiles)

# Calcular media de BMI por grupo
grupos_unicos = np.unique(grupos)
medias_bmi = [np.mean(bmi[grupos == g]) for g in grupos_unicos]

# Etiquetas para los grupos
etiquetas = ['Baja', 'Media-Baja', 'Media-Alta', 'Alta']

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
barras = plt.bar(etiquetas, medias_bmi, color=['green', 'yellow', 'orange', 'red'], 
                 alpha=0.7, edgecolor='black')

# Agregar valores en las barras
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura,
             f'{altura:.3f}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Etiquetas y título
plt.xlabel('Nivel de Progresión de la Enfermedad', fontsize=12)
plt.ylabel('BMI Promedio (normalizado)', fontsize=12)
plt.title('BMI Promedio por Grupo de Progresión de Diabetes', fontsize=14, fontweight='bold')

# Mejorar visualización
plt.grid(True, alpha=0.3, axis='y')

plt.show()
```

### Ejercicio 9

 - Cargue el dataset de diabetes.

 - Calcule la matriz de correlación para las características del conjunto de datos.

 - Cree un mapa de calor para visualizar la matriz de correlación.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes

# Cargar dataset de diabetes
diabetes = load_diabetes()
X = diabetes.data
nombres_caracteristicas = diabetes.feature_names

# Calcular matriz de correlación
matriz_correlacion = np.corrcoef(X.T)

# Crear mapa de calor
plt.figure(figsize=(12, 10))
sns.heatmap(matriz_correlacion, 
            annot=True,  # Mostrar valores
            fmt='.2f',   # Formato de 2 decimales
            cmap='coolwarm',  # Mapa de colores
            center=0,    # Centrar en 0
            square=True, # Celdas cuadradas
            linewidths=0.5,  # Líneas entre celdas
            xticklabels=nombres_caracteristicas,
            yticklabels=nombres_caracteristicas,
            cbar_kws={'label': 'Correlación'})

# Título
plt.title('Matriz de Correlación - Dataset de Diabetes', fontsize=14, fontweight='bold', pad=20)

# Rotar etiquetas para mejor lectura
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

### Ejercicio 10

 - Cargue el dataset de dígitos (load_digits) utilizando sklearn.datasets.

 - Seleccione una imagen del conjunto de datos y cree un mapa de calor para visualizarla.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits

# Cargar dataset de dígitos
digits = load_digits()
X = digits.data
y = digits.target
imagenes = digits.images

# Seleccionar una imagen (por ejemplo, la primera)
indice = 0
imagen = imagenes[indice]
etiqueta = y[indice]

# Crear mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(imagen, annot=True, fmt='.0f', cmap='gray_r', 
            cbar_kws={'label': 'Intensidad'}, square=True)

plt.title(f'Dígito: {etiqueta}', fontsize=14, fontweight='bold')
plt.xlabel('Pixel X')
plt.ylabel('Pixel Y')

plt.tight_layout()
plt.show()
```

### Ejercicio 11

 - Cargue el dataset de Boston (load_boston) utilizando sklearn.datasets.
 
 - Cree un gráfico de líneas mostrando la relación entre AGE (edad de las casas) y PRICE (precio de las casas).

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Cargar el dataset Boston
boston = fetch_openml(name="boston", version=1, as_frame=True)
df = boston.frame

# Ordenar por AGE para un gráfico más limpio
df = df.sort_values("AGE")

# Gráfico de líneas AGE vs PRICE
plt.figure(figsize=(12, 6))
plt.plot(df["AGE"], df["MEDV"], linewidth=0.5, alpha=0.7, color='steelblue')
plt.xlabel("AGE (edad de las casas)")
plt.ylabel("PRICE (precio de las casas)")
plt.title("Relación entre AGE y PRICE en el dataset Boston")

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Ejercicio 15

 - Cargue el dataset flights utilizando seaborn.load_dataset.

 - Utilice seaborn.heatmap para visualizar el número de pasajeros en función de los meses y los años.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset flights
flights = sns.load_dataset('flights')

# Pivotar datos para crear matriz mes x año
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Crear heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlOrRd', 
            linewidths=0.5, cbar_kws={'label': 'Número de Pasajeros'})

plt.xlabel('Año', fontsize=12)
plt.ylabel('Mes', fontsize=12)
plt.title('Número de Pasajeros por Mes y Año', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
