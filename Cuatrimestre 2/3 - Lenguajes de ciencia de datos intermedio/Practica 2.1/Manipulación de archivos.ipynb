# Operaciones de E/S y manipulación de archivos

## Objetivo

El objetivo de esta actividad es evaluar las habilidades en la manipulación de archivos utilizando Python. A través de una serie de ejercicios prácticos, se pondrán en práctica diversas operaciones con archivos y directorios, tales como lectura, escritura, modificación, renombrado, y eliminación.

## Instrucciones Generales

- Cada ejercicio debe ser resuelto utilizando Python utilizando programación estructurada.

- Crea un archivo Python separado para cada ejercicio, nombrándolos de manera consecutiva, por ejemplo, `ejercicio1.py`, `ejercicio2.py`, etcétera.

- Asegúrate de que cada script esté bien comentado, explicando brevemente qué hace cada parte del código.

- Guarda todos los archivos resultantes en una carpeta llamada `ManipulacionArchivosPython`.

- Prueba cada script para asegurarte de que funciona correctamente antes de pasar al siguiente ejercicio.

## Ejercicios con archivos de datos de contenido libre

1. Escribe un programa que renombre un archivo llamado `archivo1.txt` a `archivo1_renombrado.txt`.

2. Escribe un programa que elimine un archivo llamado `archivo2.txt`.

3. Crea un directorio llamado `mi_directorio`.

4. Escribe un programa que liste todos los archivos y directorios en el directorio actual.

5. Escribe un programa que copie el contenido de archivo1.txt a un nuevo archivo llamado `archivo1_copia.txt`.

6. Escribe un programa que mueva `archivo1.txt` al directorio `mi_directorio`.

7. Crea un programa que verifique si un archivo llamado `archivo1.txt` existe en el directorio actual y lo indique en pantalla.

8. Escribe un programa que lea un archivo llamado `archivo1.txt` y cuente el número de líneas que contiene.

9. Escribe un programa que lea un archivo llamado `archivo1.txt` y cuente el número de palabras que contiene.

10. Escribe un programa que lea un archivo llamado `archivo1.txt` y cuente el número de caracteres que contiene.

11. Crea un nuevo conjunto de archivos de entrada CSV en una carpeta llamada input_data. Luego, crea una carpeta de salida separada llamada output_data. Escribe un script en Python que procese múltiples archivos CSV desde la carpeta input_data, lea los archivos de entrada, procese los datos (por ejemplo, calcular totales como precio del producto por unidades vendidas) y escriba los resultados en archivos CSV en la carpeta output_data.


1. Escribe un programa en Python que lea un archivo llamado `banco_preguntas.txt`, y escriba un archivo `examen.txt` con un conjunto de cinco preguntas seleccionadas al azar. El archivo `banco_preguntas.txt` se encuentra disponible en plataforma.

2. Supongamos que tienes una hoja de cálculo con datos del Censo de EE. UU. de 2010 y la tarea tediosa de revisar miles de filas para contar tanto la población total como el número de sectores censales en cada condado. (Un sector censal es una área geográfica definida para
los propósitos del censo). Cada fila representa un único sector censal.
Aunque es posible calcular la suma de múltiples celdas seleccionadas, tendrías que seleccionar manualmente las celdas para cada uno de los más de 3,000 condados. Incluso si solo toma unos segundos calcular la población de un condado manualmente, esto tomaría horas para hacerlo en toda la hoja de cálculo.
En este proyecto, escribirás un programa que pueda leer el archivo de la hoja de cálculo del censo y calcular estadísticas para cada condado.

    - Leer los datos de la hoja de cálculo (censuspopdata.csv).

    - Contar el número de sectores censales en cada condado.

    - Contar la población total de cada condado.

    - Escribe los resultados en un archivo CSV.

3. Utiliza el archivo `winequality-both.csv` para crear dos archivos separados: uno con los registros de vino blanco y otro con los de vino tinto. Escribe un programa en Python que lea el archivo CSV, separe los registros de vino blanco (white) y tinto (red), y luego guarde estos registros en dos nuevos archivos llamados `winequality-white.csv` y `winequality-red.csv`.

4. Utiliza el archivo `TitanicSurvival.csv` para calcular estadísticas de supervivencia según la clase del pasajero. Escribe un programa en Python que lea el archivo CSV, calcule el número total de pasajeros y el número de sobrevivientes para cada clase de pasajero (Pclass), y luego calcule el porcentaje de supervivencia para cada clase. Finalmente, escriba en un archivo con formato TXT los resultados del análisis.

5. Utiliza el archivo `supplier_data.csv` para filtrar los registros de proveedores según la cantidad de productos suministrados. Escribe un programa en Python que lea el archivo CSV, filtre los proveedores que han suministrado más de una cierta cantidad de productos (por ejemplo, más de 100 productos), y luego guarde estos registros en un nuevo archivo llamado `filtered_supplier_data.csv`.

## Referencias

1. [Repositorio de GitHub con Jupiter Notebook de la Practica.](https://github.com/RKCbas/Maestria-en-Inteligencia-Artificial---Practicas)
